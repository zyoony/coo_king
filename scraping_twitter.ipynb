```python
#query ë¶€ë¶„ ê²€ìƒ‰ì–´ ë°˜ë“œì‹œ ë°”ê¿”ì£¼ê¸°
#https://twitter-michelin-guide.netlify.app/ì—ì„œ ê·¸ë£¹ ê²€ìƒ‰-ì§€ì—­ ê²€ìƒ‰ í›„ íŠ¸ìœ„í„°ì— ëœ¨ëŠ” ê²€ìƒ‰ì–´ë¡œ queryë¥¼ ë³€ê²½í•´ì£¼ì„¸ìš”

query = "í•©ì •#ëª¬ë² ë² ê°€_ëª¬ë² ë² ì—ê²Œ_ì¶”ì²œí•˜ëŠ”_ë§›ì§‘ OR #ì…”ëˆ„_ë„Œì´ë¯¸ë¨¹ì—ˆê² ì§€ë§Œë˜ë¨¹ì–´ì•¼ê² ì§€ OR #ë¯¼í˜_ë„ˆëŠ”ì´ë¯¸ë¶ë§ˆí¬í•´ë†¨ê² ì§€ OR #ê¸°í˜„_ë„ˆëŠ”ë§˜ë§ˆë¨¹ê³ ë‹¤ì‹œì‹œì‘í•˜ê² ì§€ OR #ì±„í˜•ì›_ë„ˆëŠ”ì„¸ì…ë„ëª»ë¨¹ê² ì§€  OR #ì£¼í—Œ_ë„ˆëŠ”í•œì…ê°€ë“ëª»ë¨¹ê² ì§€ OR #ì„ì°½ê· _ë„ˆëŠ”í•œë¼ëšë‹­ë¨¹ê² ì§€"
```

```python
tweet_list = []  #ë°ì´í„° í”„ë ˆì„ì„ ë§Œë“¤ê¸° ìœ„í•œ ë¹ˆ ë¦¬ìŠ¤íŠ¸ ì§€ì •
limit = 1000 #íŠ¸ìœ„í„° ê¸€ì„ 1000ê°œê¹Œì§€ë§Œ ê°€ì ¸ì˜¤ë„ë¡ ì„¤ì •

for tweet in sntwitter.TwitterSearchScraper(query).get_items():

    if len(tweet_list) == limit: #íŠ¸ìœ„í„° ê¸€ì´ 1000ê°œ ì´ìƒì¸ ê²½ìš° íŠ¸ìœ„í„° ê¸€ì„ ë” ì´ìƒ ê°€ì ¸ì˜¤ì§€ ì•ŠìŒ
        break
    else:
        tweet_list.append([tweet.date, tweet.user.username, tweet.user.displayname, tweet.content, tweet.url])
        
        ## ì´ ì½”ë“œë¡œ ì½ì–´ì˜¬ ìˆ˜ ìˆëŠ” ê²ƒ ##
        # íŠ¸ìœ„í„° ê¸€ ë‚ ì§œ
        # ìœ ì € ì•„ì´ë””
        # ìœ ì € ë‹‰ë„¤ì„
        # íŠ¸ìœ„í„° ê¸€
        # íŠ¸ìœ„í„° ë§í¬
```

```python
#ìŠ¤í¬ë˜í•‘í•œ íŠ¸ìœ„í„° ì •ë³´ë¥¼ ë°ì´í„° í”„ë ˆì„ìœ¼ë¡œ ë§Œë“¤ê¸°
df = pd.DataFrame(tweet_list, columns = ['Date', 'User', 'Nickname', 'Tweet', 'URL'])

#ë§Œë“  ë°ì´í„° í”„ë ˆì„ ë³´ì—¬ì£¼ê¸°
print(df)
```
